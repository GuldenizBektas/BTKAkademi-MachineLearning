#### A HUGE Trick

```python
data_2 = data.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
data_2
```

'tempature' and 'humidity' is a problem here. They were already numeric, but now they have been transformed anyway. So, we won't use them.

```python
c = data_2.iloc[:, :1]
```

```python
c
```

Let's apply OneHotEncoder.

```python
ohe = preprocessing.OneHotEncoder()
```

```python
c = ohe.fit_transform(c).toarray()
```

```python
final = pd.DataFrame(data=c, index=range(14), columns=['o', 'r', 's'])
```

```python
final
```

```python
df = pd.concat([final, data.iloc[:, 1:3]], axis=1)
```

```python
df
```

```python
df_final = pd.concat([data_2.iloc[:, -2:], df], axis=1)
```

```python
df_final
```

1. overcast = o
2. sunny = s
3. rainy = r

```python
from sklearn.model_selection import train_test_split
```

'humidity' is independent variable.

```python
x_train, x_test, y_train, y_test = train_test_split(df_final.iloc[:, :-1], df_final.iloc[:, -1:],
                                                   test_size=0.33, random_state=0)
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
y_test
```
